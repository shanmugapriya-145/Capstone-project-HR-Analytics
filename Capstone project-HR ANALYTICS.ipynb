```python
# Features with more than 1 missing value
missing_val_features=[each for each in df.columns if df[each].isnull().sum()>1]
print(missing_val_features)
# Bar plot showing count of missing values for each feature
for feature in missing_val_features:
    height=df[feature].isnull().sum()
    plt.bar(feature,height)
```

```python
# Total and percentage of missing data for each feature
total=df.isnull().sum().sort_values(ascending=False)
percent_1=(df.isnull().sum()/df.shape[0])*100
percent_2=round(percent_1,1).sort_values(ascending=False)
missing_data=pd.concat([total,percent_2],axis=1,keys=['Total','%'])
missing_data.head(3)
```

```python
# Impute missing values using KNNImputer
from sklearn.impute import KNNImputer
x=df[['Base_pay','openingbalance','Total_Sales']]
imputer=KNNImputer(n_neighbors=2)
x=imputer.fit_transform(x)
```

```python
# Update DataFrame with imputed values
df[['Base_pay','openingbalance','Total_Sales']]=pd.DataFrame(x,columns=['Base_pay','openingbalance','Total_Sales'])
```

# Handling Outliers

```python
# Calculate IQR
Q1=dff.quantile(0.25).round(3)
Q3=dff.quantile(0.75).round(3)
IQR=Q3-Q1
print(IQR)
```

```python
# Remove outliers
dfout=dff[~((dff < (Q1 - 1.5 * IQR)) | (dff > (Q3 + 1.5 * IQR))).any(axis=1)]
dfout.shape
```

# Model Building

```python
# Split data into features and target
x=dfout.drop(['Salary'],axis=1)
x
```

```python
y=dfout['Salary']
y
```

```python
# Split data into training and testing sets
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
# Standardize features
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

# Machine Learning Techniques

```python
#importing decisionregressor from sklearn.tree library
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor(max_depth=3)
```

```python
#fitting x_train and y_train to dt
dt.fit(x_train,y_train)
```

```python
#score of train model
print(dt.score(x_train,y_train))
#score of test model
print(dt.score(x_test,y_test))
```

Accuracy is coming 96.97% and 97.11%,which is good.

```python
#by using dt.predict function on x_test and calling it to y_predict to predict value
y_pred=dt.predict(x_test)
```

```python
print(y_pred)#checking the predicted values
```

```python
#importing metrics from sklearn library 
from sklearn import metrics
#metrics gives r2 score to cross verify accuracy of model
```

```python
#checking the r2_score to y_test,y_predict to predict the accuracy
r_square=metrics.r2_score(y_test,y_pred)
print("R-square error for decision tree regressor is:",round(100*(r_square),3))
```

```python
#calculating mae,mse,rmse,mape and accuracy on y_test
print('Mean Absolute Error(MAE):',metrics.mean_absolute_error(y_test,y1_pred))
print('Mean Squared Error(MSE):',metrics.mean_squared_error(y_test,y1_pred))
#while both MAE/MSE are performance measures
print('Root Mean Squared Error(RMSE):',np.sqrt(metrics.mean_squared_error(y_test,y1_pred)))
#RMSE is a measure of how well the machine learning model can perform.
mape=np.mean(np.abs((y_test-y1_pred)/np.abs(y1_pred)))
mape1=round(mape*100,2)
accuracy1=round(100*(1-mape),2)
print('Mean Absolute Percentage Error(MAPE):',round(mape*100,2))
print('Accuracy:',round(100*(1-mape),2))
```

# Cross validation

Cross-validation is a resampling method that uses different portions of the data to test and train a model on different iterations

```python
#Cross validation for our model
from sklearn.model_selection import ShuffleSplit
model=LinearRegression()
ssplit=ShuffleSplit(n_splits=10,test_size=0.30)
from sklearn.model_selection import cross_val_score
results=cross_val_score(model,x,y,cv=ssplit)
print(results)
print(np.mean(results))
```

As we can see the different iteration everytime, we are getting accuracy of almost 100%.So we can say that linear regression model doesnot suit our model.

```python
#Cross validation for our model
from sklearn.model_selection import ShuffleSplit
model=DecisionTreeRegressor(max_depth=3)
ssplit=ShuffleSplit(n_splits=10,test_size=0.30)
from sklearn.model_selection import cross_val_score
results=cross_val_score(model,x,y,cv=ssplit)
print(results)
print(np.mean(results))
```

For different iterations we are getting good accuracy in decision tree with th mean of 96.98% accuracy.
